```python
import pandas as pd
```

```python
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

```python
facturacion1 = tienda1["Precio"].sum()
facturacion2 = tienda2["Precio"].sum()
facturacion3 = tienda3["Precio"].sum()
facturacion4 = tienda4["Precio"].sum()

print("Facturación Tienda 1:", facturacion1)
print("Facturación Tienda 2:", facturacion2)
print("Facturación Tienda 3:", facturacion3)
print("Facturación Tienda 4:", facturacion4)
```

```python
print("Ventas por categoría - Tienda 1")
print(tienda1["Categoría del Producto"].value_counts())

print("\nVentas por categoría - Tienda 2")
print(tienda2["Categoría del Producto"].value_counts())

print("\nVentas por categoría - Tienda 3")
print(tienda3["Categoría del Producto"].value_counts())

print("\nVentas por categoría - Tienda 4")
print(tienda4["Categoría del Producto"].value_counts())
```

```python
print("Calificación promedio Tienda 1:", tienda1["Calificación"].mean())
print("Calificación promedio Tienda 2:", tienda2["Calificación"].mean())
print("Calificación promedio Tienda 3:", tienda3["Calificación"].mean())
print("Calificación promedio Tienda 4:", tienda4["Calificación"].mean())
```

```python
print("Productos más vendidos - Tienda 1")
print(tienda1["Producto"].value_counts().head())

print("\nProductos más vendidos - Tienda 2")
print(tienda2["Producto"].value_counts().head())

print("\nProductos más vendidos - Tienda 3")
print(tienda3["Producto"].value_counts().head())

print("\nProductos más vendidos - Tienda 4")
print(tienda4["Producto"].value_counts().head())
```

```python
print("Costo de envío promedio Tienda 1:", tienda1["Costo de envío"].mean())
print("Costo de envío promedio Tienda 2:", tienda2["Costo de envío"].mean())
print("Costo de envío promedio Tienda 3:", tienda3["Costo de envío"].mean())
print("Costo de envío promedio Tienda 4:", tienda4["Costo de envío"].mean())
```
